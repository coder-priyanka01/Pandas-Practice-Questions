## I/O (Input/Output) 

```python
import pandas as pd
```

```python
# 131. Read from CSV: Ek CSV file (students.csv) ko ek DataFrame mein read karein. 
df = pd.read_csv('students.csv')
df
```

```python
# 132. Write to CSV: Ek DataFrame ko output.csv naam ki file mein save karein. 
data = {
    'StudentID': [1, 2, 3],
    'Name': ['Priyanka', 'Rahul', 'Anita'],
    'Age': [20, 21, 19],
    'Grade': ['A', 'B', 'A']
}

df = pd.DataFrame(data)
df.to_csv('output.csv', index=False) 
```

```python
# 133. Write to CSV without index: DataFrame ko ek CSV file mein save karein, lekin index column ko shamil na karein.
df.to_csv('output.csv', index=False)
df
```

```python
# 134. Read an Excel file: Ek Excel file (data.xlsx) ki pehli sheet ko ek DataFrame mein read karein.
pd.read_excel('data.xlsx', sheet_name=0)
```

```python
# 135. Write to an Excel file: Ek DataFrame ko output.xlsx naam ki file mein Sheet1 par save karein.
df.to_excel('output.xlsx', sheet_name='Sheet1', index=False)
df
```

```python
# 136. Read a specific sheet from Excel: Excel file ki Grades naam ki sheet ko read karein.
pd.read_excel('output.xlsx', sheet_name='Sheet1')
```

```python
# 137. Read a JSON file: Ek JSON file (data.json) ko ek DataFrame mein read karein.
pd.read_json('data.json')
```

```python
# 138. Write to a JSON file: Ek DataFrame ko output.json file mein save karein.
df.to_json('output.json', orient='records')
df
```

```python
# 139. Read data from clipboard: Apne clipboard se table-formatted data ko seedha DataFrame mein read karein.
df = pd.read_clipboard()
df
```
