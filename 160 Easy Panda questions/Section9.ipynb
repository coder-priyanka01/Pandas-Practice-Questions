##  Basic Plotting 

```python
import pandas as pd
import numpy as np
import matplotlib as plt

# Base DataFrame
data = {
    'Name': ['Alice', np.nan, 'Charlie', 'David', 'Eva', 'Mayra',' Priyanka'],
    'Age': [25, 30, 35, 40, 22, 18, 19],
    'City': ['New York', 'London', 'Paris', 'New York', np.nan, 'Moscow', 'Berlin'],
    'Salary': [50000, np.nan, 55000, 70000, 65000, 72000, 80000],
    'Neg_col' : [-12, -34, -45, -56, -67 ,-78, -89]
}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6','row7'])
print(df)
```

```python
# 121. Line Plot: Ek numerical column (e.g., 'Stock_Price') ka ek simple line plot banayein.
df['Salary'].plot.line()
```

```python
# 122. Bar Plot: 'City' column ke value counts ka ek bar plot banayein. 
df['City'].value_counts().plot.bar()
```

```python
# 123. Histogram: 'Age' column ka distribution dekhne ke liye ek histogram banayein. 
df['Age'].hist(bins=6, color='skyblue', edgecolor='black', figsize=(8,5), grid=False)
```

```python
# 124. Box Plot: 'Salary' column ka ek box plot banayein taaki quartiles aur outliers dekh sakein.
df['Salary'].plot.box()
```

```python
# 125. Area Plot: Do numerical columns ka ek stacked area plot banayein. 
df[['Age','Neg_col']].plot.area(stacked=True)
```

```python
# 126. Scatter Plot: Do numerical columns (e.g., 'Age' vs 'Salary') ke beech relationship dekhne ke liye ek scatter plot banayein.
df.plot.scatter(x='Age', y='Salary')
```

```python
# 127. Pie Chart: 'Gender' column ke distribution ko represent karne ke liye ek pie chart banayein. 
df['City'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# 128. Plot with Title: Ek line plot banayein aur uska title "Stock Price Over Time" set karein.
df['Salary'].plot(title='Stock Price Over Time')
```

```python
# 129. Plot with Labels: Ek bar plot banayein aur uske X-axis ko 'City' aur Y-axis ko 'Count' label dein.
city_counts = df['City'].value_counts()
city_counts.plot(kind='bar', xlabel='City', ylabel='Count', title='Count of Records per City')
```

```python
# 130. Horizontal Bar Plot: 'City' column ke value counts ka ek horizontal bar plot (barh) banayein.
df['City'].value_counts().plot.barh()
```
