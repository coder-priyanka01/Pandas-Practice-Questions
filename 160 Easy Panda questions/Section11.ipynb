##  Combining DataFrames 

```python
import pandas as pd
```

```python
# 141. Concatenate two DataFrames (rows): Do alag-alag DataFrames ko vertically (ek ke neeche ek) concatenate karein.
df1 = pd.read_csv("output.csv")
df2 = pd.read_csv("File.csv")

df = pd.concat([df1, df2], axis=0)
df
```

```python
# 142. Concatenate two DataFrames (columns): Do DataFrames ko horizontally (side-by-side) concatenate karein.
df1 = pd.read_csv("output.csv")
df2 = pd.read_csv("File.csv")

df = pd.concat([df1, df2], axis=1)
df
```

```python
# 143. Append a row: Ek DataFrame mein ek nayi row (dictionary ya Series ke roop mein) append karein.
df = pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
new_row = {'col1': 5, 'col2': 6}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

print(df)
```

```python
# 144. Append a DataFrame: Ek DataFrame ko doosre DataFrame mein append karein. 
df = pd.read_csv("output.csv")
other_df = pd.read_csv("File.csv")

df = pd.concat([df, other_df], ignore_index=False)
df
```

```python
# 145. Concatenate with ignore_index: Do DataFrames ko concatenate karein aur unke index ko reset karne ke liye ignore_index=True ka use karein.
df1 = pd.read_csv("output.csv")
df2 = pd.read_csv("File.csv")
pd.concat([df1,df2], ignore_index=True)
```

```python
# 146. Concatenate with keys: Do DataFrames ko concatenate karein aur unke source ko track karne ke liye keys add karein.
df1 = pd.read_csv("output.csv")
df2 = pd.read_csv("File.csv")
pd.concat([df1,df2], keys=['src1','src2'])
```

```python
# 147. Join two DataFrames on their indices: Do DataFrames ko unke index par join karein.
df1 = pd.read_csv("output.csv", index_col="Age")
df2 = pd.read_csv("File.csv", index_col="Age")

df = df1.join(df2)  
df
```

```python
# 148. Merge two DataFrames (inner join): Do DataFrames ko ek common column par inner join ka use karke merge karein.
df1 = pd.read_csv("output.csv")
df2 = pd.read_csv("File.csv")
pd.merge(df1, df2, on='Age', how='inner')
```

```python
# 149. Simple merge: pd.merge() ka use karke do DataFrames ko ek 'ID' column par merge karein.
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Bina', 'Chetan']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Age': [25, 30, 35]
})

df =  pd.merge(df1, df2, on='ID')
df
```

```python
# 150. Append a Series to a DataFrame: Ek Series ko DataFrame ki nayi row ke roop mein append karein.
df = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Amit', 'Bina', 'Chetan']
})
new_row = pd.Series({'ID': 4, 'Name': 'Deepa'})
(new_row)
df.loc[len(df)] = new_row
df
```
