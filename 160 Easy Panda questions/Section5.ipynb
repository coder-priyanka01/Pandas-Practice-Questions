##  Basic Data Manipulation 

```python
import pandas as pd
import numpy as np

# Base DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Mayra',' Priyanka'],
    'Age': [25, 30, 35, 40, 22, 18, 19],
    'City': ['New York', 'London', 'Paris', 'New York', 'Tokyo', 'Moscow', 'Berlin'],
    'Salary': [50000, 60000, 55000, 70000, 65000, 72000, 80000]
}
df = pd.DataFrame(data, index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6','row7'])
print(df)
```

```python
# 41. Add a new column: DataFrame mein ek 'Salary' column add karein aur sabhi rows ke liye value 50000 set karein.
df['Salary']=50000
df
```

```python
#42. Add a column based on existing columns: Ek 'Age_in_Months' column banayein jo 'Age' column ko 12 se multiply karke calculate ho.
df['Age_in_Months']=df['Age']*12
df
```

```python
# 43. Set a specific value: loc ka use karke, index 'row1' aur column 'Salary' ki value ko 60000 set karein.
df.loc['row1','Salary']=60000
df
```

```python
# 44. Set values for an entire row: iloc ka use karke, pehli row ke sabhi values ko 0 set karein.
df.iloc[0]=0
df
```

```python
#45. Drop a column: DataFrame se 'Salary' column ko drop() method ka use karke hatayein.
df.drop(columns=['Salary'])
```

```python
# 46. Drop multiple columns: DataFrame se do columns ko ek saath hatayein. 
df.drop(columns=['Age','Age_in_Months'])
df
```

```python
# 47. Drop a row by index label: loc ka use karke, index label 'row2' waali row ko hatayein.
df.drop(index=['row2'])
```

```python
# 48. Drop rows by index position: iloc ka use karke, index position 0 par maujood row ko hatayein.
df.drop(df.index[0])
```

```python
# 49. Drop duplicate rows: DataFrame se sabhi duplicate rows ko hatayein. 
df.drop_duplicates()
```

```python
# 50. Rename a single column: Column 'Name' ka naam badal kar 'Full_Name' karein. 
df.rename(columns={'Name':'Full_Name'})
```

```python
# 51. Rename multiple columns: Ek dictionary ka use karke 'Age' ko 'Years' aur 'City' ko 'Location' mein rename karein. 
df.rename(columns={'Age':'Years','City':'Location'})
```

```python
# 52. Apply a function to a column: apply() method ka use karke 'Age' column ke har element mein 1 add karein.
df['Age']=df['Age'].apply(lambda x: x+1)
df
```

```python
# 53. Replace values: 'City' column mein jahan bhi 'New York' likha hai, use 'NYC' se replace karein. 
df['City']=df['City'].replace('New York','NYC')
df
```

```python
# 54. Change a column's data type: 'Salary' column ka data type float se int mein badlein.
df['Salary']=df['Salary'].astype(int)
df
```

```python
# 55. Set a value for a slice of rows: Index 1 se 3 tak ki rows ke liye 'Salary' column ki value 75000 set karein. 
df.iloc[1:4, df.columns.get_loc('Salary')] = 75000
df
```

```python
# 56. Extract a sample of rows: DataFrame se randomly 3 rows ka sample nikal kar display karein.
df.sample(3)
```

```python
# 57. Create a new column with a default value: Ek 'Is_Employee' column banayein aur uski default value True set karein.
df['Is_Employee']=True
df
```

```python
# 58. Modify a column in-place: 'Age' column ke sabhi values ko 2 se divide karein.
df['Age']=df['Age']/2
df
```

```python
# 59. Combine two string columns: 'First_Name' aur 'Last_Name' columns ko jod kar ek 'Full_Name' column banayein.
df_names = pd.DataFrame({
    'First_Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Last_Name': ['Smith', 'Johnson', 'Brown', 'Davis', 'Miller']
})
df_names['Full_Name'] = df_names['First_Name'] + ' ' + df_names['Last_Name']

print(df_names)
```

```python
# 60. Reset the index and drop the old one: DataFrame ka index reset karein aur purane index ko ek column ke roop mein add na karein. 
df.reset_index(drop=True)
```
