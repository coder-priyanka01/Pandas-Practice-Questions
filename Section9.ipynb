## : Advanced apply, pipe, and Performance 

```python
import pandas as pd
import numpy as np
data = { 
'Math_Score': [88, 92, 85, 76, 95, 70], 
'Science_Score': [90, 85, 93, 80, 88, 75], 
'History_Score': [78, 88, 94, 82, 85, 79] 
} 
df_scores = pd.DataFrame(data)
df_scores
```

```python
df_emp = pd.DataFrame({
    'Department': ['HR','IT','Finance','IT','HR','Finance','IT','HR'],
    'Employee': ['Alice','Bob','Charlie','David','Eve','Frank','Grace','Heidi'],
    'Salary': [70000,80000,90000,85000,72000,95000,88000,75000],
    'Experience': [5,8,10,7,6,12,9,7]
})
```

```python
# 91. Write a function that takes a row of df_scores as input and returns 'Pass' if the average score is >= 85, and 'Fail' otherwise. Apply this function across the rows (axis=1) to create a new 'Result' column. 
def result(row):
    return 'Pass' if row.mean() >= 85 else 'Fail'
df_scores['Result'] = df_scores.apply(result, axis=1)
df_scores
```

```python
# 92. Using df_emp (from the first set), use groupby() on 'Department' and then .transform() to create a 'Dept_Avg_Salary' column. Explain why transform is useful here. 
df_emp['Dept_Avg_Salary'] = df_emp.groupby('Department')['Salary'].transform('mean')
df_emp
```

```python
# 93. Define two separate functions: one that adds 5 to all scores and one that caps all scores at 100. Use the .pipe() method to apply both functions sequentially to df_scores. 
def add5(df): return df + 5
def cap100(df): return df.clip(upper=100)
df_scores_pipe = df_scores[['Math_Score','Science_Score','History_Score']].pipe(add5).pipe(cap100)
df_scores_pipe
```

```python
# 94. Calculate the standard deviation of scores for each student (row-wise) in df_scores. 
df_scores[['Math_Score','Science_Score','History_Score']].std(axis=1)
```

```python
# 95. Use the applymap() function to format every score in df_scores into a string like "Score: 88". 
df_scores[['Math_Score','Science_Score','History_Score']].applymap(lambda x: f"Score: {x}")
```

```python
# 96. Create a DataFrame with object-type columns that contain numbers stored as strings. Write the code to identify and convert these columns to a numeric type efficiently. 
df_obj = pd.DataFrame({'col1':['1','2','3'], 'col2':['4','5','6']})
df_obj = df_obj.apply(pd.to_numeric, errors='coerce')
df_obj.dtypes
```

```python
# 97. From df_sales, apply a function to the 'Sales' column that gives a 10% bonus if sales are over 200 and a 5% bonus otherwise. 
df_sales = pd.DataFrame({
    'Region': ['North','South','East','West'],
    'Sales': [150, 250, 180, 300]
})
df_sales['Bonus'] = df_sales['Sales'].apply(lambda x: x*0.10 if x>200 else x*0.05)
df_sales
```

```python
# 98. In df_emp, create a new column 'Salary_to_Experience_Ratio' by applying a lambda function row-wise. 
df_emp['Salary_to_Experience_Ratio'] = df_emp.apply(lambda r: r['Salary']/r['Experience'], axis=1)
df_emp[['Employee','Salary_to_Experience_Ratio']]
```

```python
# 99. Create a function that calculates the interquartile range (IQR) of a series. Apply this function to each column of df_scores. 
def IQR(series):
    return series.quantile(0.75) - series.quantile(0.25)
df_scores[['Math_Score','Science_Score','History_Score']].apply(IQR)
```

```python
# 100. Using df_emp, create a new column 'Salary_Group' that is a copy of the 'Salary' column, but downcast it to the most memory-efficient integer type possible.
df_emp['Salary_Group'] = pd.to_numeric(df_emp['Salary'], downcast='integer')
df_emp[['Employee','Salary_Group']].dtypes
```
